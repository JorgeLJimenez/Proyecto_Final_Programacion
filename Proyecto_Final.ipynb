```python
''' nombre del programa: PROCESAMIENTO DE DATOS TERMODINÁMICOS USANDO PYTHON Y MYSQL.

autor: Est. Jorge Luis Jiménez Hurtado
cod est.: 1005704849
institución: Universidad del Quindío
correo institucional: jorgel.jimenezh@uqvirtual.edu.co

Buen día. Por favor empieza a leer el programa desde la última línea y te guiarás por las llamadas de las funciones.
'''
import os
from pathlib import Path
#from sqlalchemy import create_engine
#import pymysql
#import sys
import mysql.connector 
import random
import pandas as pd
import xlrd
import re # la uso en el método to_relate()
import math # La uso en el método thermo_process()
import numpy as np
#from numpy.linalg import eig, LinAlgError


'''Procesamiento de tablas de estados y procesos termodinámicos dentro de bases de datos usando MySQL.'''
#
#
#Este método me muestra el menú principal (main menu).
def user_options():
    while True:
        print("\nMAIN MENU\n(1)Create database\n(2)Delete database\n(3)Login to a database\n(4)Exit.\n\nChoose an option (1/2/3/4/5)")
        response = input('>>>')     
        if response == '1':
            create_db()       
        elif response == '2':
            delete_db()
        elif response == '3':
            login_db()
        elif response == '4':
            raise Exception('Has cerrado el programa.')
        else:
            input_error()
#
#
# Cada vez que el usuario no ingrese un dato válido, se llamará a esta función.
def input_error():
    print("\nRespuesta no válida.\n")
#
#
# Método para ingresar el nombre de una tabla, mostrándome las existentes en la base de datos.
def input_table(conn, cursor):
    # Retrieving the list of tables. (retrieve - n. recuperar.)
    print("List of tables: ")
    cursor.execute("SHOW TABLES")
        
    for x in cursor:
        print(x)

    print("Ingrese el nombre de la tabla")
    tb=input('>>>')
    
    return tb
#
#
#
def connect_and_show_db():
    # Establishing the connection. Good practice: use 'conn' or 'cnx'
    conn = mysql.connector.connect(
        host="localhost",
        user="root",
        passwd="Himenesumusic_13",
    )
    # Creating a cursor object using the cursor() method
    cursor = conn.cursor()
    
    # Retrieving the list of databases. (retrieve - n. recuperar.)
    print("List of databases: ")
    cursor.execute("SHOW DATABASES")
    
    for x in cursor:
        print(x)
    return conn, cursor
#
#
'''Creating a database in MySQL using python.'''
def create_db():
    conn,cursor = connect_and_show_db()
    MyNewDB=input('Ingresa el nombre de la nueva base de datos. ' 
                  'PELIGRO: Si el nombre de la base de datos que vas a crear coincide con alguna existente, '
                  'entonces se reemplazará la existente por la nueva.\n>>>')
    
    # Dropping a database deletes the database from an instance of SQL Server and deletes the physical disk files used by the database.    
    # Dropping database MyDB_Thermo if already exists. DROP database is a keyword of MySQL.
    cursor.execute("DROP database IF EXISTS " + db_name)

    # Preparing query to create a database. CREATE database is a keyword of MySQL.
    sql = "CREATE database " + db_name

    # Creating a database. MySQLCursor.execute() Method executes the given database operation (query(consulta o acción sobre los datos) or command).
    cursor.execute(sql)

    # Retrieving the list of databases. (retrieve - n. recuperar.)
    print("List of databases: ")
    cursor.execute("SHOW DATABASES")
    
    for x in cursor:
        print(x)

    print(cursor.fetchall()) # MySQLCursor.fetchall() Method fetches(fetch - v. ir a buscar) all (or all remaining) rows of a query result set and returns a list of tuples. If no more rows are available, it returns an empty list.
    
    # Comando que me guarda los cambio en MySQL y para que se muestren en MySQL Workbench
    conn.commit()
    
    # Closing the connection
    conn.close()
    
    pass
#
# Este método me elimina una base de datos existente elegida por el usuario.
def delete_db():
    cnx,cursor = connect_and_show_db()
    
    ByeDB=input("Ingresa el nombre de la base de datos que desea eliminar.\n>>>")
    
    # Dropping a database deletes the database from an instance of SQL Server and deletes the physical disk files used by the database.    
    # Dropping database MyDB_Thermo if already exists. DROP database is a keyword of MySQL.
    cursor.execute("DROP database IF EXISTS " + db_name)
    
    # Retrieving the list of databases. (retrieve - n. recuperar.)
    print("List of databases: ")
    cursor.execute("SHOW DATABASES")
    
    for x in cursor:
        print(x)

    #print(cursor.fetchall()) #MySQLCursor.fetchall() Method fetches(fetch - v. ir a buscar) all (or all remaining) rows of a query result set and returns a list of tuples. If no more rows are available, it returns an empty list.
    
    # Comando que me guarda los cambio en MySQL y para que se muestren en MySQL Workbench
    cnx.commit()
    
    # Closing the connection
    cnx.close()
    
    pass
#
#
def login_db():
    cnx,cursor = connect_and_show_db()        
    cnx.close()
    # While trying to connect to an existing database.
    while not cnx.is_connected():
        toDB=input("Ingrese el nombre de la base de datos a la cual desea entrar.\n>>>")
        try:
            # Establishing the connection. Good practice: use 'conn' or 'cnx'  
            cnx = mysql.connector.connect(
            host="localhost",
            user="root",
            passwd="Himenesumusic_13",
            database=toDB
            )
            print("Conexión exitosa.\n")
            cursor = cnx.cursor()
            processing_data_menu(cnx,cursor)
        except mysql.connector.Error:
            print(f"No fue posible conectarse porque la base de datos '{toDB}' no existe.\n")
        else: print("Error al intentar conectarse.\n")
            
    pass
#
#
def processing_data_menu(conn,cursor):
    obj=Process_data('tb_states77.xlsx','tb_states78.xlsx','tb_states79.xlsx','test_table137.xlsx',
                     tb_states77=r'C:\Users\Windows 10\Desktop\UNIQUINDÍO\5to Semestre\Proyecto_Final',
                     data_base1=r'C:\Users\Windows 10\Desktop\UNIQUINDÍO\5to Semestre\Proyecto_Final\OtroDirectorio',
                     tb_states78=r'C:\Users\Windows 10\Desktop\UNIQUINDÍO\5to Semestre\Proyecto_Final',
                     test_table137=r'C:\Users\Windows 10\Desktop\UNIQUINDÍO\5to Semestre'
                    )
    while True:
        print("\nMAIN MENU\n(1)Usar operaciones CRUD de MySQL\n(2)Procesar datos\n(3)Exit\nChoose an option (1/2/3)")
        response = input('>>>')     
        if response == '1':
            submenu_tables_mysql_options(conn,cursor,obj)      
        elif response == '2':
            processing_data(conn,cursor)
        elif response == '3':
            user_options()
        else:
            input_error()       
#
# Menú de opciones de tablas disponibles para el usuario con MySQL
def submenu_tables_mysql_options(conn,cursor,obj):
    while True:
        print("\nSUBMENU\n(1)Create table\n(2)Read table\n(3)Update table\n(4)Delete table\n(5)Exit.\n\nChoose an option (1/2/3/4/5)\n")
        response = input('>>>')
        
        if response == '1':
            obj.create_table(conn, cursor, obj)
        elif response == '2':
            obj.read_table(conn, cursor)
        elif response == '3':
            obj.update_table(conn, cursor)
        elif response == '4':
            obj.delete_table(conn, cursor, obj)
        elif response == '5':
            processing_data_menu(conn, cursor)
        else:
            input_error()
    pass
#
# 
def read_my_cwdir():
    '''We get the name of my current working directory as a string'''
    cwdir = os.getcwd()
    '''We get files and directories in my current working directory in a list'''
    files_in_cwdir = os.listdir() 
    '''Para obtener el directorio de trabajo actual, usamos el función getcwd() (get current working directory) del módulo os. 
    Esta función te devuelve el directorio actual en forma de cadena.'''
    '''I get the files of xlsx type in a list'''
    valid_files = [i for i in files_in_cwdir if os.path.isfile(i) and i.endswith('.xlsx')]
    
    print(f"\nDirectorio de trabajo actual: {cwdir}")
    print(f"Archivos y carpetas dentro: {files_in_cwdir}.")
    print(f"Archivos válidos (de tipo .xlsx): {valid_files}")
        
    return cwdir,files_in_cwdir,valid_files
#
#
'''Mi decorador :-)'''
def decorator_check_cwdir(function_to_decorate, pa, kwa): # pa: positional arguments, kwa: keyword arguments
    cwdir, files_in_cwdir, valid_files = read_my_cwdir()
    
    def is_cwdir(path):
        if path == cwdir: 
            return 'actual '
        else:
            return ''
    print("")    
    for i,j in zip(kwa.keys(),kwa.values()): # zip() me permite iterar con más de un índice. Python’s zip function allows us to loop over multiple lists at the same time:
        path = Path(j+'\\'+i+'.xlsx') # ejemplo cuando concateno: r'C:\Users\Windows 10\Desktop\UNIQUINDÍO\5to Semestre\Parcial2\data_base3.xlsx'
        if path.is_file():
            print(f"El archivo {i+'.xlsx'} existe en el directorio {is_cwdir(j)}{j}")
        else:
            print(f"El archivo {i+'.xlsx'} no existe en el directorio {is_cwdir(j)}{j}")
    
    function_to_decorate()
    pass
#
# Clase que al inicializarse recibe como argumentos los nombres de archivos y opcionalmente sus direcciones.
class Process_data():
    #
    '''Al crearse la clase recibe nombres de archivos y opcionalmente sus direcciones'''
    def __init__(self, *args, **kwargs): # args=(var1,var2,var3,...,varn), kwargs={a:'val1', b:'val2',...}
        decorator_check_cwdir(self.__check_my_cwdir__, args, kwargs)
    #
    #
    '''aplico la función decoradora 'decorator_check_actdir()' a la función de abajo'''
    def __check_my_cwdir__(self):
        print("Se ha inicializado la clase Process_data() y se han leído los archivos recibidos.")
    #
    '''Los siguientes métodos de la clase Process_data() son la implementación de las operaciones CRUD de MySQL.'''
    #
    #
    def create_table(self, conn, cursor,obj):
        print("Ingrese el nombre de la tabla P,V,T que desea crear")
        statesTBname=input('>>>')
        
        #Dropping newTB table if already exists.
        cursor.execute(f"DROP TABLE IF EXISTS {statesTBname}")
        
        #Creating table as per requirement
        sql = f"CREATE TABLE {statesTBname} (state INT, P FLOAT(7,2), V FLOAT(7,2), T FLOAT(7,2))" #Here, (M,D) means than values can be stored with up to M digits in total, of which D digits may be after the decimal point
        
        try:
            cursor.execute(sql)
        except mysql.connector.Error as err:
            print("Failed creating database: {0}".format(err))
            exit(1)
            
        print("Ingrese el nombre de la tabla de procesos asociada a la tabla de estados")
        processesTBname=input('>>>')
         
        #Dropping newTB table if already exists.
        cursor.execute(f"DROP TABLE IF EXISTS {processesTBname}")
        
        #Creating table as per requirement
        sql = f"CREATE TABLE {processesTBname} (initial_final_states VARCHAR(255), type_of_process VARCHAR(255))"
        
        try:
            cursor.execute(sql)
        except mysql.connector.Error as err:
            print("Failed creating database: {0}".format(err))
            exit(1)

        # Preparing SQL query to INSERT a record into the database.
        insDynSta = f"INSERT INTO {statesTBname}(state, P, V, T) VALUES (%s, %s, %s, %s)" # %s se usar para insertar los valores dinámicamente.
        insDynPro = f"INSERT INTO {processesTBname}(initial_final_states, type_of_process) VALUES (%s, %s)"
        
        while True:
            print("(1)Enter values\n(2)Random values\n(3)Exit\nPlease choose an option (1/2/3)")
            resp=input(">>>")
            
            if resp == '1':
                self.insert_in_table(conn, cursor, insDynSta, insDynPro)
                submenu_tables_mysql_options(conn,cursor,obj)
            elif resp == '2':
                self.random_table(conn, cursor, insDynSta, insDynPro)
                submenu_tables_mysql_options(conn,cursor,obj)
            elif resp == '3':
                submenu_tables_mysql_options(conn,cursor,obj)
            else:
                input_error()

        pass
    #
    #
    #
    # Método que permite al usuario insertar uno o varios registros en una tabla cuyo nombre está guardado en la variable tb
    def insert_in_table(self, conn, cursor, insDynSta, insDynPro):
        state=0 # flag - bandera
        resp='y'
        while resp == 'y':
            state+=1
            tonteando=True     
            while tonteando:
                print(f"Ingrese la presión, volumen, y temperatura del estado {state}")
                P=input("P = ")
                V=input("V = ")
                T=input("T = ")              
                try:
                    P=float(P)
                    V=float(V)
                    T=float(T)
                    if P>0 and V>0 and T>0:
                        tonteando=False
                    else:
                        print("Solo se permiten números reales mayores a cero.")
                except:
                    print("El dato ingresado no es de tipo flotante.")
            try:
                # Populating the table 
                cursor.execute(insDynSta,(state,P, V, T))
                # Commit your changes in the database
                conn.commit()
            except:
                # Rolling back in case of error (roll back - v. retroceder (una acción))
                conn.rollback()
                print("Error al intentar ejecutar query. Los datos a insertar en la tabla no son de tipo flotante.")
            if state > 1:
                ups=True
                while ups:
                    print(f"Ingrese el tipo de proceso llevado a cabo del estado {state-1} al estado {state}")
                    proc=input(">>>")
                    if proc in ["adiabatic","isothermal","isobaric","isochoric"]:
                        ups=False 
                        trans=f'{state-1}-{state}'
                        cursor.execute(insDynPro,(trans,proc))
                        conn.commit()
                    else:
                        input_error()
            while True:
                resp = input("Desea ingresar un nuevo estado (P,V,T) ? y/n\n>>>")
                if resp in ['y','n']:
                    break
                else:
                    input_error()
    
        pass
    #
    #
    #   
    def random_table(self, conn, cursor, insDynSta, insDynPro):
        tonteando=True
        while tonteando:
            n=input("Ingrese el número de estados (número de filas de la tabla)\n>>>")
            try:
                n = int(n)
                if n > 1:
                    tonteando=False
                else:
                    print("Solo se permiten enteros mayores o iguales a dos.")
            except ValueError:
                print("El dato ingresado no es de tipo entero.")

        for i in range(n):
            P,V,T = random.uniform(0, 100),random.uniform(0, 100),random.uniform(0, 100)
            # Populating the table 
            cursor.execute(insDynSta,(i+1,P,V,T))
            
        for i in range(n-1):
            trans=f'{i+1}-{i+2}'
            proce=random.choice(["adiabatic","isothermal","isobaric","isochoric"])
            cursor.execute(insDynPro,(trans,proce))
            
        # Commit your changes in the database
        conn.commit()  
        
        pass
    #
    #
    #
    # Método para leer, selecionar información útil de la base de datos a la cual el usuario está conectado (conn).
    def read_table(self, conn, cursor):
        tb = input_table(conn, cursor) 
        #Retrieving single row
        sql = f'SELECT * from {tb}'
        
        #Executing the query
        cursor.execute(sql)
        
        print("Todos los registros de la tabla:")
        #Fetching 1st row from the table
        result = cursor.fetchall();
        print(result)
        
        '''print("El primer registro de la tabla:")
        #Fetching 1st row from the table
        result = cursor.fetchone();
        print(result)'''
        
        print("Registros con T > 3:")
        #Retrieving specific records using the where clause
        #cursor.execute(f"SELECT * from {tb} WHERE 373.25 < T < 373.25") #temperaturas entre 0 y 100 grados centigrados.
        cursor.execute(f"SELECT * from {tb} WHERE T > 3") 
        print(cursor.fetchall())
        
        print("Registros ordenados por V ascendiente:")
        cursor.execute(f"SELECT * from {tb} ORDER BY V") # ordena los registros por V ascendiente.
        print(cursor.fetchall())
        
        print("Registros ordenados por P descendiente:")
        cursor.execute(f"SELECT * from {tb} ORDER BY P DESC") # ordena los registros por P descendiente.
        print(cursor.fetchall())
        
        pass
    #
    #
    #   
    def update_table(self, conn, cursor):
        tb = input_table(conn, cursor) 
        #Preparing the query to update the records
        sql_query = f'''UPDATE {tb} SET P = 10, V = 20, T = 30 WHERE P = V AND V = T'''
        
        try:
            # Execute the SQL command
            cursor.execute(sql_query)
            # Commit your changes in the database
            conn.commit()
        except:
            # Rollback in case there is any error
            conn.rollback()
            print("Error al intentar modificar el registro.")

        pass
    #
    #
    # metodo que me permite eliminar registros de una tabla
    def delete_table(self, conn, cursor, obj):
        
        while True:
            print("(1)Delete a record of the table\n(2)Drop table\n(3)Exit\nChoose an option (1/2/3)")
            resp=input(">>>")
            
            if resp == '1':
                self.delete_record(conn, cursor)
            elif resp == '2':
                self.drop_table(conn, cursor)
            elif resp == '3':
                submenu_tables_mysql_options(conn,cursor,obj)
            else:
                input_error()
        
        print("Eliminar tabla")
        
        self.drop_table(conn, cursor)
    #
    #
    #
    '''Método que me elimina registros según condición dada sobre los datos.'''
    def delete_record(self, conn, cursor):
        tb = input_table(conn, cursor)
        #Preparing the query to delete records
        sql = f"DELETE FROM {tb} WHERE P <= 0 OR V <= 0 OR T <= 0 OR ( P = V AND V = T )"
        try:
            # Execute the SQL command
            cursor.execute(sql)
            # Commit your changes in the database
            conn.commit()
        except:
            # Roll back in case there is any error
            print("Hubo un error en la ejecución para eliminar los registros de la tabla.")
            conn.rollback()
        pass
    #
    #
    #
    '''Método que me elimina una tabla guardada en una base de datos usando MySQL'''
    def drop_table(self, conn, cursor):
        tb = input_table(conn, cursor)
        #Doping tb table if already exists
        try: 
            cursor.execute(f"DROP TABLE {tb}")
            print(f"Table {tb} dropped... ")
        except:
            # Roll back in case there is any error
            print("Hubo un error en la ejecución para eliminar la tabla.")
            conn.rollback()
        pass
#
#
'''Menú para el procesamiento de datos: aquí se llaman varias funciones de procesan las dataframes con datos termodinámicos'''
def processing_data(conn,cursor):
    alright=False
    dfs1={} # voy a guardar todas mis tablas de mysql en el diccionario dfs1. {'name_table1':'table_data_frame1',...}
    dfs2={} # voy a guardar todas mis tablas de mi directorio en el diccionario dfs2. {'name_table2':'table_data_frame2',...}      
    while not alright:
        print("(1)Importar datos de mysql a pandas dataframe\n",
            '(2)Importar datos de mi directorio actual a pandas dataframe\n',
            '''(3)¡Estoy listo para procesar mis datos!''')
        resp=input(">>>")
        subresp='y'
        if resp=='1':
            while subresp=='y':
                df_name,df = imp_mysql_df(conn,cursor)
                #dfs = {**dfs, df_name:df} # adding a key:value pair using a dictionary comprehension.
                dfs1.update({df_name:df}) # adding a key:value pair using the update() method.
                print("Desea importar otra tabla de MySQL?\n(1)y\n(2)n")
                subresp=input('>>>')                    
        elif resp=='2':
            while subresp=='y':
                df_name,df = import_from_my_cwdir()
                #dfs = {**dfs, df_name:df} # adding a key:value pair using a dictionary comprehension.
                dfs2.update({df_name:df}) # adding a key:value pair using the update() method.
                print("Desea importar otra tabla del directorio actual?\n(1)y\n(2)n")
                subresp=input('>>>')
        elif resp=='3':
            alright=True
        else:
            input_error()
            
    dfs={**dfs1,**dfs2} # junto todas las tablas obtenidas de mysql y del directorio actual, identificas con su nombre por defecto. ¡¡¡No puedo juntar tablas con el mismo nombre (key)!!!
    
    print(dfs)
    
    print("Tienes varias tablas para procesar. Elige qué acción quieres efectuar sobre ellas:")    
    
    while True:
        print("(1)Calcular calor y trabajo total en cada tabla\n",
             "(2)Resolver un sistema AX=Y\n",
             "(3)Mostrar un gráfico de regresión lineal con scatter y un gráfico de caja y bigote\n",
             "(4)Ingresar función de dos variables y graficar, dadas dos columnas elegida por el usuario alguna tabla. (funcion lambda)",
             "(5)Exit")
        resp=input(">>>")
        
        if resp=='1':
            compute_heatwork(dfs)
        elif resp=='2':
            solve_matrix_equation(dfs)
        elif resp=='3':
            plot_linear_regression(dfs)
            plot_caja_bigote(dfs)
        elif resp=='4':
            plot_function(dfs)
        elif resp=='5':
            processing_data_menu(conn,cursor)
        else:
            input_error()
            
    # pendiente...
#
#
'''Summary of this method: importing data from a MySQL database into a Pandas data frame including column names'''
def imp_mysql_df(conn,cursor):
    goout=False
    while not goout:
        try:
            tb=input_table(conn, cursor)
            cursor.execute(f"SELECT * FROM {tb}")
            L1=cursor.fetchall()
            '''Once data is fetched it can be loaded into DataFrame'''
            df=pd.DataFrame(L1) # con este comando no obtengo los registros sin nombres de columnas.
            cursor.execute(f"SHOW COLUMNS FROM {tb}")
            L2=cursor.fetchall() 
            #df_column_names=pd.DataFrame(cursor.fetchall())            
            goout=True
        except mysql.connector.Error as err:
            print("Failed selecting table '{0}': {1}\n".format(tb,err))
    print(L1)
    print(df)
    print(L2)
    L=[]
    for i in range(len(L2)): 
        L.append(L2[i][0])
    print(L)
    # adding column name to the respective nameless columns
    df.columns=L
    print(df)
    print(f'''\nSe ha importado desde mysql la tabla {tb} a dataframe con éxito.\nNueva dataframe {tb}:''')
    display(df)
            
    return tb,df

'''Método para importar un archivo de tipo .xlsx de mi directorio de trabajo actual. Retorna una tabla de tipo dataframe'''
def import_from_my_cwdir():
    cwdir,files_in_cwdir,valid_files = read_my_cwdir()
    While=True
    while While:
        print("\nIngrese el nombre de la tabla con terminación .xlsx que quiere importar del directorio actual a pandas dataframe\n")
        tb=input('''>>>''')
        if tb in valid_files:
            While=False
            df=pd.read_excel(tb)
            display(df)
        else: 
            print("El archivo no existe en el directorio.")
            
    return tb.replace('.xlsx',''),df
#
#
'''Método que me permite calcular el calor y el trabajo en cada transformación y el total. Se muestra al usuario'''
def compute_heatwork(dfs):
    ordered=to_relate(dfs) # e.g. ordered=[[dfs1, dfp1], [dfs3, dfp3], [dfs2, dfp2], [dfs1, dfp1], [dfs3, dfp3], [dfs2, dfp2]]'''
    
    print("\nCalor y trabajo en cada transformación:")
    for i in ordered:
        df_stat=i[0] # i-ésima tabla dataframe de estados
        df_proc=i[1] # i-ésima tabla dataframe de procesos
        df_QW = pd.DataFrame(columns=['initial_final_states','Q','W'])
        '''obtengo las columnas de las dataframes como listas'''
        statelis=df_stat['state'].tolist()
        Plis=df_stat['P'].tolist()
        Vlis=df_stat['V'].tolist()
        Tlis=df_stat['T'].tolist()
        ifslis=df_proc['initial_final_states'].tolist()
        typlis=df_proc['type_of_process'].tolist()
        Q_T,W_T=0,0
        for j in range(len(typlis)):
            Q,W = thermo_process(j+1,typlis[j],Plis[j],Plis[j+1],Vlis[j],Vlis[j+1],Tlis[j],Tlis[j+1])
            print(f"{ifslis[j]}: Q = {round(Q,2)} joules, W= {round(W,2)} joules")
            entry = pd.DataFrame.from_dict({
                "initial_final_states":[ifslis[j]],
                "Q":[round(Q,2)],
                "W":[round(W,2)]
            })
            df_QW=pd.concat([df_QW, entry], ignore_index=True)
            Q_T+=Q
            W_T+=W
        display(df_QW)
        print(f"Calor total: {round(Q_T)} joules, Trabajo total: {round(W_T)} joules")
    pass
#
#
'''La función to_relate() me ayuda a hacer corresponder las tablas de los estados y procesos'''
'''EJEMPLO:
Si la funcion recibe
dfs={'tb_processes1':dfp1,'tb_processes3':dfp3,'tb_states2':dfs2,'tb_states1':dfs1,'tb_states2':dfs2,'tb_states3':dfs3,'tb_processes2':dfp2}
donde dfs1,dfs2,...,dfp1, dfp2... son las dataframes que quiero relacionar en pares, entonces la relacion es la lista que retorna la funcion:
ordered=[[dfs1, dfp1], [dfs3, dfp3], [dfs2, dfp2], [dfs1, dfp1], [dfs3, dfp3], [dfs2, dfp2]]'''
def to_relate(dfs):
    ordered=[]
    for x,y in zip(dfs.keys(),dfs.values()):
        i=re.findall(r'\d',x)[0] # este metodo me retorna una lista de strings con los números del string original x, y me paro en el indice 0 para obtener el primer string(número). En este caso particular, retorna una lista de un solo elemento)
        for u,w in zip(dfs.keys(),dfs.values()):
            j=re.findall(r'\d',u)[0]
            if j==i and x!=u: # si los nombres tienen el mismo número de terminación pero no son iguales (dfs1 != dfs1).
                # Para agregar primero tabla de estados, luego tabla de procesos.
                if x.find('states') != -1: # si x es el nombre de la tabla de estados. (find() retorna la posición índice del primer carácter del string 'states' en el string x. Retorna -1 si no lo encuentra)
                    ordered.append([y,w])
                else: # sino x es el nombre de la tabla de procesos.
                    ordered.append([w,y])
                break # no se va a encontrar un tercer nombre con igual número de terminación. No siga buscando/iterando.
    n=int(len(ordered)/2) 
    ordered=ordered[:n] # la segunda mitad de la lista se repite, debo cortarla (by using a slice).
    print(ordered)
    return ordered

'''Método que me retorna el calor y el trabajo realizado del estado 'state' al estado 'state+1'.'''
def thermo_process(state,process,P1,P2,V1,V2,T1,T2):
    R=8.314
    if process=="adiabatic":
        Q, W = 0, -3*R*(T2-T1)
    elif process=="isobaric":
        Q, W = 5*R*(T2-T1), P1*(V2-V1)
    elif process=="isochoric":
        Q, W = 3*R*(T2-T1), 0
    elif process=="isothermal":
        Q, W = 2*R*T1*math.log(V2/V1,math.e), 2*R*T1*math.log(V2/V1,math.e)
    return Q, W

def solve_matrix_equation(dfs):
    '''agrego las dataframes de estados state,P,V,T convertidas a listas en una lista.'''
    '''recordar: Sean dfsi,dfpi, i=1,2,3... las dataframes, dfs es el diccionario a procesar. e.g:
    dfs={ 'tb_processes1':dfp1,'tb_processes3':dfp3,'tb_states2':dfs2,'tb_states1':dfs1,
        'tb_states2':dfs2,'tb_states3':dfs3,'tb_processes2':dfp2 }'''
    listdata=[dfs[i].values.tolist() for i in dfs.keys() if i.find('states') != -1]
    #print(str(listdata).replace(']],',']]\n\n'))
    
    for i in range(len(listdata)):
        obj=Matrix(listdata[i])
        print("\nMatriz a procesar:")
        print(obj) # usando el método especial __str__()
        #print(np.array(obj.M))
        M=obj.M
        m=obj.shape[0]
        n=obj.shape[1]
        print(f"Tamaño de matriz: {m}x{n}")
        print("\nMatrix transpuesta:")
        obj_T=Matrix(obj.__transpose__())
        print(obj_T)

        if m < n: # m=2, m=3. pues n>3 (state,P,V,T)
            obj2=SquaredMatrix(convert_to_squared(M,m,n,)) # agrego una fila con valores aleatorios.
            print("\nMatriz convertida a cuadrada:")
            print(obj2) # usando el método especial __str__() 
        elif m > n: 
            obj2=SquaredMatrix(convert_to_squared(M,m,n,m-n)) # # agrego m-n columnas con valores aleatorios.
            print("\nMatriz convertida a cuadrada:")
            print(obj2) # usando el método especial __str__()
        elif m==n: 
            print("\nLa matriz ya es cuadrada!")
            print(obj)
        print(f"Tamaño de nueva matriz: {obj2.shape[0]}x{obj2.shape[1]}")
        obj2_T=Matrix(obj2.__transpose__())
        print("\nMatriz cuadrada transpuesta:")
        print(obj2_T)
        print("\nDeterminante:")
        print(obj2.__determinant__())
        print("\nInversa:")
        print(obj2.__inverse__())
        print("\nAutovalores y autovectores:")
        print(obj2.__eigen__())
        b=[[random.uniform(-100, 100)] for i in range(m)]
        print(f"\nLa solución del sistema Ax=b, donde\nb = {b}\nA = {obj2.M}\nes")
        print(f"x = {obj2.__solve__(b)}")
        
        print("\nPROCESAMIENTO DE LA SIGUIENTE TABLA DE DATOS\n")
#   
# 
'''Mis tres decoradores que se aplican sobre el método especial __str__(). Estos darán un formato parecido al del array de pandas'''
'''Mi primer decorador: me convierte la lista a un string'''
def convert_to_string(function_to_decorate):
    def wrapper(ref): # el ref es necesario cuando la función a decorar está dentro de una clase.
        func=function_to_decorate(ref)
        return str(func)
    return wrapper
'''Mi segundo decorador: me reemplaza las ], por ]\n. Es decir, visualizo las filas una sobre otra.'''
def get_matrix_form(function_to_decorate):
    def wrapper(ref):
        func=function_to_decorate(ref)
        return func.replace('],',']\n')
    return wrapper
'''Mi tercer decorador: me reemplaza las coma por un espacio. Así se parecerá más a la notación matemática'''
def replace_comma(function_to_decorate):
    def wrapper(ref):
        func=function_to_decorate(ref)
        return func.replace(',',' ')
    return wrapper
#
#
'''superclase Matrix'''
class Matrix:
    def __init__(self,listdf):
        self.M=listdf
        self.shape=self.__get_shape__(listdf)
        pass
    @replace_comma
    @get_matrix_form
    @convert_to_string
    def __str__(self):
        return self.M # intentando emular como se imprimiría un numpy.ndarray
    '''sin los decoradores, solo pondría el comando:
        return str(self.M).replace('],',']\n').replace(',',' ')'''
    def __get_shape__(self,L):
        return [len(L),len(L[0])] # [m,n]
    def __add__(self,otra):
        return [list(map(lambda x,y:x+y,self.M[k],otra[k])) for k in range(self.shape[0])]
    def __transpose__(self):
        T=[]
        for j in range(self.shape[1]): T+=[[self.M[i][j] for i in range(self.shape[0])]]
        return T

'''subclase de mi superclase Matrix'''
class SquaredMatrix(Matrix):
    def __init__(self,Mx):
        m,n=len(Mx),len(Mx[0])
        '''super() gives you access to attributes and methods in a superclass from the subclass that inherits from it'''
        if m==n: # si es cuadrada...
            super().__init__(Mx)
        else:
            super().__init__(convert_to_squared(Mx,m,n,m-n)) # sino convertir a cuadrada.
    def __determinant__(self):
        return np.linalg.det(np.array(self.M))
    def __inverse__(self):
        try:
            return np.linalg.inv(np.array(self.M)).tolist()
        except LinAlgError:
            print("Singular matrix. La matrix no tiene inversa.")
            return [[]]
    def __eigen__(self):
        eigenval,eigenvec=eig(np.array(self.M))
        return eigenval.tolist(),eigenvec.tolist()
    def __solve__(self,b): # resuelve el sistema Ax=b
        try:
            x=np.linalg.solve(np.array(self.M),np.array(b))
            return x.tolist()
        except LinAlgError:
            print("Sistema matricial sin solución. La matriz A es singular.")

def convert_to_squared(M,m,n,l=None):
    '''np.random.randint(arg1,arg2,[m,n]) me da un objeto de tipo 'numpy.ndarray'. Si lo convierto a lista,
    obtengo una lista de m-filas por n-columnas con valores aleatorios en el intervalo entre arg1 y arg2.
    '''
    if l!=None:
        return [M[j]+[round(random.uniform(0.01,100),1) for i in range(l)] for j in range(m)] # agrego l columnas aleatorias.
    else:
        return M+[[round(random.uniform(0.01,100),1) for i in range(n)] for j in range(n-m)] # agrego una fila aleatoria.

'''Otras funciones buenas para procesar los datos. No se han implementado por el momento...'''
def plot_linear_regression():
    pass
def plot_caja_bigote():
    pass
def plot_input_function():
    pass

'''ESTA FUNCIÓN ME COMIENZA A EJECUTAR TODO EL PROGRAMA. ES MI MAIN.'''
user_options()
```
